# Ted topics: separating and classifying

```python
type(ted["topics"][0])
```

```python
# column"topics": each item is a string with lots of keywords. let's make a list with all the strings:
topics_list = []
for i in range(len(ted["topics"])):
    topic = ted["topics"][i].split(",")
    topics_list.append(topic)
    
    
```

```python
type(ted["topics"][0])
```

```python
topics_list[0][2]
```

```python
# let's put this list of lists in one single list of keywords
single_t_list = []
for i in topics_list:
    for x in i:
        single_t_list.append(x)
        
```

```python
#how many keywords do I have?
len(single_t_list)
```

```python
# clean the list: get rid of single quotes, spaces, square brackets:
# use strip to clean spaces
for i in range(len(single_t_list)):
    single_t_list[i] = single_t_list[i].replace("[","")
    single_t_list[i] = single_t_list[i].replace("'","")
    single_t_list[i] = single_t_list[i].replace("]","")
    single_t_list[i] = single_t_list[i].strip()
```

```python
# how many unique values do we have:
len(set(single_t_list))
```

```python
# how many values do we have for each keyword
list_of_topics = pd.Series(single_t_list)
```

```python
#in percentage
list_of_topics.value_counts(normalize=True)
```

```python
#in absolute values
list_of_topics.value_counts()
```

```python
# getting a more complete idea of keywords, to see which could be the ultimate topic groups
list_of_topics_df = list_of_topics.value_counts().to_frame()
list_of_topics_df.head(50)
```

## creating one column for each group and dummifying observations

```python
type(ted["topics"][0])
```

```python
import re
```

```python
#making a function that finds all of the therms i defined as technology
#it takes a list and returns 0 if nothing is found, and 1 if one or more terms are found
technology = re.compile(r"technology|innovation|invention|computers|engineering")
def tech_topic(topics):
    if technology.findall(topics) == []:
        return 0
    else:
        return 1 
```

```python
#applying function to all the rows, and storing result in a new column called technology
ted["technology"] = ted["topics"].apply(tech_topic)
```

```python
ted["technology"].value_counts(normalize=True)
```

```python
#doing the same and creating 5 columns- one for each group
science = re.compile(r"science|future")
def science_topic(topics):
    if science.findall(topics) == []:
        return 0
    else:
        return 1 

ted["science&future"] = ted["topics"].apply(science_topic)
```
